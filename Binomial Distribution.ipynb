```python
import math
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
plt.style.use('ggplot')
```

```python
p=0.3
N=100

probs={}
for n in range(1,N+1):
    level1={}
    for x in range(1,n+1):
        factorial=math.factorial(n)/(math.factorial(x)*math.factorial(n-x))
        probability=factorial*(p**x)*((1-p)**(n-x))
        level1[x]=probability
    probs[n]=level1

n=50
probs[n]

keys=[k for k,v in probs[n].items()]
values=[v for k,v in probs[n].items()]
cumsum=0
cumsum_list=[]
for v in values:
    cumsum=cumsum+v
    cumsum_list.append(cumsum)
    
#cumsum_list2=[stats.binom.cdf(k,n,p) for k,v in zip(keys,values)] 
    
    

fig,ax=plt.subplots(figsize=(10,5))
ax.bar(keys,values,label='Probability')
ax.set_title(f'PDF AND CDF (n={n}, p={p})',fontweight='bold')
ax.legend(loc='upper left',fontsize=8)
ax.set_ylabel('Probability')
#ax.set_yscale('log')
ax2=ax.twinx()
ax2.plot(keys,cumsum_list,color='b',linewidth=1,label='Cumulative Probability')
ax2.legend(loc='upper right',fontsize=8)
ax2.set_ylabel('Cumulative Probability')

ax.grid(axis='both')
plt.show()
```
