```python
import pandas as pd
pd.options.display.max_columns=100
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
weights1=np.arange(0.01,1.01,0.01)
weights2=np.arange(0.01,1.01,0.01)
```

```python
# 2 weights

w1s=[]
w2s=[]

for w1 in weights1:
    for w2 in weights2:
        if w1+w2==1:
            w1s.append(w1)
            w2s.append(w2)
```

```python
P=0.2

M1=0.1
STD1=0.16
M2=0.15
STD2=0.24
```

```python
MPs=[]
for w1,w2 in zip(w1s,w2s):
    MP=w1*M1+w2*M2
    MPs.append(MP)
```

```python
STDPs=[]
for w1,w2 in zip(w1s,w2s):
    STDP=(w1**2*STD1**2+w2**2*STD2**2+2*w1*w2*STD1*STD2*P)**0.5
    STDPs.append(STDP)
```

```python
plt.scatter(x=STDPs,y=MPs,s=5)
plt.title('Risk vs Expected Return',fontweight='bold')
plt.show()
```

# Insurance 

```python
data=pd.read_html('https://www.ssa.gov/OACT/STATS/table4c6.html')[0]
```

```python
df=data.iloc[:-1,:]
```

```python
df=df.apply(pd.to_numeric)
```

```python
plt.figure(figsize=(15,6))
plt.plot(df['Exact  age'].iloc[:,0].astype(int),df['Male']['Death  probability a'],label='Male')
plt.plot(df['Exact  age'].iloc[:,0].astype(int),df['Female']['Death  probability a'],label='Female')
plt.fill_between(df['Exact  age'].iloc[:,0].astype(int),df['Male']['Death  probability a'],df['Female']['Death  probability a'],alpha=0.3)
plt.title('Death probability by age')
plt.legend()
plt.show()
```
